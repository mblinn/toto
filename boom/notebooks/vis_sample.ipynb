```
from dataset_utils import download_boom_benchmark

boom_path = '/content'
dataset_dir = download_boom_benchmark(boom_path)
print('Dataset downloaded to', dataset_dir)
```

```
import json
from gift_eval.data import Dataset
import matplotlib.pyplot as plt

props = json.load(open('boom/boom_properties.json'))
example_names = list(props.keys())[:3]

for name in example_names:
    term = props[name]['term']
    ds = Dataset(name=name, term=term, to_univariate=False, storage_env_var='BOOM')
    first = next(iter(ds.test_data))['target']
    if len(first.shape) == 2:
        first = first[0]
    plt.figure(figsize=(12,3))
    plt.plot(first)
    plt.title(f'{name} - {term}')
    plt.show()
```
